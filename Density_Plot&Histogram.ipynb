```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
df = pd.read_csv("1500000 Sales Records (1).csv")
df.head(5)
```

```python
# Plotting the density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Units Sold'], fill=True, color='skyblue', alpha=0.5)
plt.title('Density Plot of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Density')
plt.xlim(0, df['Units Sold'].max() + 10)  # Adjust x-axis limits for better visualization
plt.grid()

# Show the plot
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Units Sold'], bins=15, kde=True)
plt.title('Histogram of Units Sold')
plt.show()
```
