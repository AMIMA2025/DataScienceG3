```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("1500000 Sales Records (1).csv")

# Selected columns 
columns_of_interest = ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
selected_data = df[columns_of_interest]

# correlation matrix
corr_matrix = selected_data.corr(method='pearson')

# Heatmap visualization
plt.figure(figsize=(12, 8)) 
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Sales Features")
plt.show()

print("Correlation Matrix:")
print(corr_matrix)
```
