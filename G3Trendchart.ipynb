```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
#Loading data
file_path = "1500000 Sales Records (1).csv"
df = pd.read_csv(file_path)

# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')  # Convert to datetime,

#plotting trend charts(Total Profit by order date)
plt.figure(figsize=(10, 6))
df.groupby(df['Order Date'].dt.year)['Total Profit'].sum().plot(kind='line', marker='o', color='black')
plt.title('Total Profit by Order Date')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
#plotting trend charts(Units sold by order priority)
plt.figure(figsize=(10, 6))
df.groupby(df['Order Priority'])['Units Sold'].sum().plot(kind='line', marker='o', color='black')
plt.title('Units Sold by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```
