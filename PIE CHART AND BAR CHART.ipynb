```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
df = pd.read_csv("1500000 Sales Records (1).csv")
df.head(5)
```

```python
# Group by Order Priority and sum the Revenue
revenue_distribution = df.groupby('Order Priority')['Total Revenue'].sum().reset_index()

# Sort values for better visualization
revenue_distribution = revenue_distribution.sort_values(by='Total Revenue')

# Plotting using Seaborn
plt.figure(figsize=(7, 6))
sns.barplot(x='Order Priority', y='Total Revenue', data=revenue_distribution, hue='Order Priority', palette='viridis', legend=False)
plt.title('Distribution of Revenue by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
df['Order Priority'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
