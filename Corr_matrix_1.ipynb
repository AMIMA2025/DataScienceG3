```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\OMEN X\Desktop\New folder (2)\DataScienceG3\1500000_Sales_Records.csv")
```

```python
df.head()
```

```python
corr_columns = ["Units Sold", "Unit Price", "Unit Cost", "Total Revenue", "Total Cost", "Total Profit"]
corr_df=df[corr_columns]
corr_df.head()
```

```python
# Compute correlation matrix
corr_matrix = corr_df.corr()

# Visualizing correlation with heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True,cmap="coolwarm",  fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
